```python

status_counts = df['STATUS'].value_counts()
colors = ['red', 'blue', 'green', 'orange', 'purple'] 

plt.bar(status_counts.index, status_counts.values,color = colors)
plt.xlabel('Status')
plt.ylabel('Count')
plt.title('Status Counts')
plt.show()
```

```python
status_counts = df['PRODUCTLINE'].value_counts()
colors = ['red', 'blue', 'green', 'orange', 'purple'] 
plt.figure(figsize=(15, 5))

plt.bar(status_counts.index, status_counts.values,color = colors)
plt.xlabel('productline')
plt.ylabel('Count')
plt.title('Status Counts')
plt.show()
```

```python
status_counts = df['DEALSIZE'].value_counts()
colors = ['red', 'blue', 'green', 'orange', 'purple']
plt.pie(status_counts, labels=status_counts.index, colors=colors)
plt.title('DEL SIZE Distribution')
plt.axis('equal')
plt.show()
```
